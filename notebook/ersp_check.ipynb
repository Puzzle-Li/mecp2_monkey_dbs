# PSD

说明一下，我对如何处理不等长的数据的freq domain不太清楚，所以为了使得频率等长，我取了不同的fft参数，如下。这样的做法是否会产生错误，请查实。

```python
psd_bs = epochs_bs.compute_psd(fmax=40, n_fft=512, n_per_seg=512 // 4, method='welch')
psd_dr = epochs_dr.compute_psd(fmax=40, n_fft=512, n_per_seg=256 // 4, method='welch')
psd_bs_df = psd_bs.to_data_frame()
psd_dr_df = psd_dr.to_data_frame()
```

```python
psd_bs_df['condition'] = 0
```

```python
psd_df = pd.concat([psd_bs_df, psd_dr_df])
```

```python
from scipy.stats import ttest_ind, levene

pvalues = np.zeros((50))
for i, freq in enumerate(psd_bs.freqs):
    A = psd_bs.get_data()[:, 0, i]
    B = psd_dr.get_data()[:, 0, i]
    _, p_square = levene(A, B)
    res = ttest_ind(A, B, equal_var=p_square > 0.05)
    pvalues[i] = res.pvalue
    
```

```python
sns.lineplot(
    x="freq",
    y="14-12-14-13",
    hue="condition",
    data=psd_df
)
pvalues_plot = np.zeros(50)
pvalues_plot[:] = np.nan
pvalues_plot[pvalues< 0.05] = 0.016
plt.plot(psd_bs.freqs, pvalues_plot, 'r')
plt.show()
```
